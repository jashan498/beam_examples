```python
from sympy import Symbol, symbols, init_printing
from sympy.physics.continuum_mechanics.beam import Beam
init_printing()
```

```python
E, I = symbols('E I')
b = Beam(20, E, I)
R1, R2 = symbols('R1, R2')
w = Symbol('w')
b.apply_load(R1, 0, -1)
b.apply_load(R2, 20, -1)
b.apply_load(-w, 0, 0, end=20)
b.bc_deflection = [(0, 0),(20, 0)]
b.solve_for_reaction_loads(R1, R2)
```

```python
b.reaction_loads
```

```python
b.plot_shear_force(subs={w:80})
```

```python
b.plot_bending_moment(subs={w:80})
```

```python
b.plot_slope(subs={w:80, E:100, I:200})
```

```python
b.plot_deflection(subs={w:80, E:100, I:200})
```

```python
b.plot_loading_results(subs={w:80, E:100, I:200})
```
