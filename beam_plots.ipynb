```python
# A beam of length 20 meters is fixed at both of its ends. A point load
# 'F' is applied in the middle.
from sympy import Symbol, symbols, init_printing
from sympy.physics.continuum_mechanics.beam import Beam
init_printing()
```

```python
E, I = symbols('E I')
b = Beam(20, E, I)
M1, M2 = symbols('M1, M2')
R1, R2 = symbols('R1, R2')
F = Symbol('F')
b.apply_load(R1, 0, -1)
b.apply_load(M1, 0, -2)
b.apply_load(R2, 20, -1)
b.apply_load(M2, 20, -2)
b.apply_load(-F, 10, -1)
b.bc_deflection = [(0, 0),(20, 0)]
b.bc_slope = [(0, 0),(20, 0)]
b.solve_for_reaction_loads(R1, R2, M1, M2)
b.deflection()
```

```python
b.plot_shear_force(var={F:10})
```

```python
b.plot_bending_moment(var={F:10})
```

```python
b.plot_slope(var={F:10, E:20, I:40})    # `var` to substitute variables
```

```python
b.plot_deflection(var={F:10, E:20, I:40})
```

```python
b.plot_all(var={F:10, E:20, I:40})
```
